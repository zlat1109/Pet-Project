```python
soup = BeautifulSoup(file, 'lxml')
table1 = soup.find('table', class_='sinborde')
```

```python
table1
```

```python
headers = []
for i in table1.find_all(class_='Estilo2')[0:6]:
    title = i.text
    headers.append(title)
```

```python
headers
```

```python
school_info = pd.DataFrame(columns = headers)
```

```python
school_info
```

```python
for j in table1.find_all('tr')[1:2110]:
    row_data = j.find_all('td')
    row = [i.text for i in row_data]
    length = len(school_info)
    school_info.loc[length] = row
```

```python
school_info = school_info.drop([0], axis=0).reset_index(drop=True)
```

```python
links = soup.find_all(class_='Estilo1')


link_urls = []
for link in soup.find_all('a'):
    link_urls.append(link.get('href'))
```

```python
school_info['link_urls'] = link_urls[0:2108]
```

```python
lines = list(school_info['link_urls'])

data_dict = []
count = 0

for line in lines:
    q = requests.get(line, verify=False)
    result = q.content

    try:
        soup = BeautifulSoup(result, 'lxml')
        latitude = soup.find_all(class_='Estilo1')[-2].text
        longitude = soup.find_all(class_='Estilo1')[-1].text
        
        images = soup.find_all(class_='nivelCentro')[0].find_all('img')[5:-3]
        school_inf = []
        for image in images:
            school_inf.append(image['title'])
    except:
        np.nan
        
    data = {
            'latitude': latitude,
            'longitude': longitude,
            'school_inf' : school_inf
           }
    

    data_dict.append(data)
#person_name_company = person.strip().split(',')
#person_name = person_name_company[0]
#person_company = person_name_company[1].strip()
```
